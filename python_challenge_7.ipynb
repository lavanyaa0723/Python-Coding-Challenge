Python coding challenge 7 (day43)-While Loop, else, break, continue

```python
#Randomly choose a number between 1 and 20.

import random
secret_number = random.randint(1, 20)
max_attempts = 5

#Let the user guess the number, allowing up to 5 attempts.
print("Welcome to the number guessing game")
print("Think of a number between 1 and 20")
print("You have 5 attempts to guess it")

# Loop for 5 attempts
for attempt in range(1, max_attempts + 1):
#After each guess, tell them if their guess is too high, too low, or correct.
#If the guess is correct, break the loop and congratulate the player.

    guess = int(input(" Enter your guess: "))
    if guess == secret_number:
        print("Congratulations! You guessed the correct number!")
        break
    elif guess < secret_number:
        print("Too low, try again.")
    else:
        print("Too high, try again.")
else:
    print(" Sorry, you have used all 5 attempts.")
    print(f"The secret number is: {secret_number}")
```
