FEEDBACK ANALYSER SCENARIO

```python
#Ask the user to input a feedback sentence
feedback = input("Please enter a feedback : ")
```

```python
# count how many words in the sentence
words = feedback.split()
word_count = len(words)
print ("number of words in the sentence:",word_count)
```

```python
#how many time the word "data" appears
data_count = feedback.count("Data")
print ("number of times the word 'data' appears:",data_count)
```

```python
#Extract and print the first 5 and last 5 characters
first_five_characters = feedback[:5]
last_five_characters = feedback[-5:]
print ("first 5 characters:",first_five_characters)
print ("last 5 characters:",last_five_characters)
```

```python
# Reverse the Sentence and print
reversed_sentence = feedback[::-1]
print ("reversed sentence:",reversed_sentence)
```
