```python
#Create a tuple with fixed product categories like Electronics, Stationery, and Clothing.
Products_Category = ("Electronics", "Stationery", "Clothing")
print(Products_Category)
```

```python
#Create a list of products that can be updated dynamically.
products = ["Laptop", "Notebook", "T-Shirt"]
```

```python
print("fixed products Chategory:",Products_Category)
print("Initial Products:", products)
```

```python
# 3. Ask the user to enter a new product and add it to the list
new_product = input("Enter a new product to add: ")
products.append(new_product)
print("Updated Products List:", products)
```

```python
# 4. Ask the user to input the index of a product to remove
remove_index = int(input("Enter the index of the product to remove: "))
if 0 <= remove_index < len(products):
    removed_product = products.pop(remove_index)
    print(f"Removed product: {removed_product}")
else:
    print("Invalid index. No product removed.")

    print("Updated Products List:", products)
```

```python
# 5. Display final result
print("Final Products List:", products)
print("Product Categories (Fixed):", Products_Category)
```
