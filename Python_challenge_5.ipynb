Python coding challenge 5-Dictionary, Set, Frozenset

```python
#Create a dictionary with student names as keys and their grades as values
student_grades = {
    "Rahul": 87,
    "Lavanyaa": 99,
    "Madhu": 95,
    "Sneha": 80,
    "Nithya": 90 }

print(student_grades)
```

```python
#Allow the user to add or update grades for at least 3 students.
print("--Updated Grades--")
for i in range(3):
  name = input("Enter student name: ")
  grade = int(input("Enter student grade: "))
  student_grades[name] = grade
print(student_grades)
```

```python
#Create a set of hobbies and allow the user to input at least 3 hobbies.
hobbies = set()
print(" Enter 3 hobbies")
for i in range(3):
    hobby = input("Enter a hobby: ")
    hobbies.add(hobby)
print("--Your hobbies--:")
print(hobbies)
```

```python
# Convert the set of hobbies into a frozenset and display it.
frozen_hobbies = frozenset(hobbies)
print(frozen_hobbies)
```
