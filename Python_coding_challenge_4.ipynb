```python
#Create a 2D list where each row holds a contactâ€™s name, phone number, and email address.
contacts =[["Lavanyaa","8903523081","lavanyaa@0723.com"],["Madhu","8890401382","madhu@011.com"],["Priya","9834269261","priya@201.com"],["Kamali","9705247713","kamali@123.com"]]
print(contacts)
```

```python
#Allow the user to add a new contact.
name = input("Enter the name: ")
phone = input("Enter the phone number: ")
email = input("Enter the email address: ")
new_contact = [name, phone, email]
contacts.append(new_contact)
print("contacts added successfully")
print(contacts)
```

```python
#Display all contacts sorted alphabetically by name.
contacts.sort(key=lambda x: x[0])
print("Contacts sorted alphabetically by name:")
for contact in contacts:
    print(contact)
```

```python
#Allow the user to search for contacts by name. The search should support partial matches.
search_name = input("Enter the name to search: ").lower()
for contact in contacts:
    if search_name in contact[0].lower():
        print(contact)

```

```python
#After sorting, display only the first 3 contacts in the list.
print("First 3 contacts after sorting:\n")
for contact in contacts[:3]:
    print(contact)
```
