```python
#Ask the user to enter their weight (in kilograms) and height (in meters).
weight=float(input("Enter your weight in kilograms:"))
height=float(input("Enter your height in meters:"))
print("weight =",weight)
print("height =",height)
```

```python
#Calculate their BMI using the formula BMI = weight / (height Ã— height).
BMI=weight/(height*height)
print("BMI =",BMI)
```

```python
#Classify the BMI as underweight, normal, overweight, or obese using conditional statements.
if BMI<18.5:
    print("Underweight")
elif BMI>=18.5 and BMI<=24.9:
    print("Normal")
elif BMI>=25 and BMI<=29.9:
    print("Overweight")
else:
    print("Obese")
```

```python
#Display the result clearly.
print("--- BMI RESULT ---")
print(f"Weight: {weight} kg")
print(f"Height: {height} m")
print(f"Your BMI is: {round(BMI, 2)}")
```
